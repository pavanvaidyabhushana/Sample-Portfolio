```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.catplot(data = data , x = "Sex", kind = "count").set(title="Count of Male&Female")
```

```python
sns.catplot(data = data, kind = "count" , x = "Survived").set(title = "Number of Survived")
```

```python
sns.catplot(data = data, kind = "count" , x = "Sex" , hue = "Survived").set(title = "Number of Survived with Sex data")
```

```python
sns.catplot(data = data, kind = "count" , x = "Pclass" , hue = "Survived").set(title = "Number of Survived with Sex data")
```

```python
data["Agegroup"] = pd.qcut(data["Age"],4)
sns.barplot(data = data , y = "Survived" , x = "Agegroup").set(title = "Survived wrt Age groups")
```

```python
sns.catplot(data = data , x = "Pclass" , col= "Embarked" ,kind = "count" ,hue = "Survived")
```

```python
sns.catplot(data = data , col = "Embarked" , hue = "Sex" , kind = "count" , x = "Survived").set(
    title = "No of male/female survived at Embarked position")
```

```python
Family = data["Parch"] + data["SibSp"]
data = data.assign(Family_size = Family)
data
```

```python
sns.barplot(data= data , x = "Family_size" , y = "Survived").set(
title = "% of ppl survived in each family depends on size")
```

```python
data.loc[data['Family_size'] == 0,'FamilySize'] = "Alone"
data.loc[data['Family_size'] != 0,'FamilySize'] = "Family"
sns.catplot(data = data , x = "FamilySize" , kind = "count" , hue = "Survived").set(
    title = "Survived ")
```

```python
sns.violinplot(data = data,x='Sex',y='Age',hue='Survived')
```

```python
group = data.groupby(['Pclass','Survived'])
pclass_survived = group.size().unstack()
sns.heatmap(pclass_survived,fmt='d',annot=True)
```
